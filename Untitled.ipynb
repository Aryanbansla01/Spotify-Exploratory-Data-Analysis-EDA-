```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

## Setting up the Enviroment to handle E Notation

```python
def fmt(x): #Setting up the Enviroment to handle E Notation
    return '%.2f' % x
pd.set_option('display.float_format',fmt)
```

## Loading DataSet

```python
df_tracks=pd.read_csv("D:/Data Sets/spotify/tracks.csv")
df_tracks.head(8)
```

## Checking Null Values

```python
pd.isnull(df_tracks).sum()
```

df_tracks.info()

## Statistical View of Data

```python
df_tracks.describe()
```

```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

## Most Listening song

```python
top_10=df_tracks.groupby(by="artists").duration.sum().sort_values(ascending=False).head(10)
top_10
plt.figure(figsize=(20,10))
plt.barh(y=top_10.index,width=top_10.values)
for y,width in zip (top_10.index,top_10.values):
    plt.text(width,y,width+1,size=16)
plt.title("10 Most Listening song",size=50)
plt.ylabel('Name of song',size=25,color='r')
plt.xlabel('Duration',size=25,color='r')
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()
```

## Top 10 Popular Song

```python
top_popularity=df_tracks.set_index('name',inplace=False)
top_pop=top_popularity.sort_values(by='popularity',ascending=False)
top_pop.popularity.head(5)
```

```python
plt.figure(figsize=(15,10))
plt.barh(y=top_pop.popularity.head(10).index,width=top_pop.popularity.head(10).values,color=('g','r','b','y','pink','g','r','b','y'))
for y,width in zip (top_pop.popularity.head(10).index,top_pop.popularity.head(10).values):
    plt.text(width,y,width+1,size=16)
plt.title("10 Popular Song",size=50)
plt.ylabel('Name Of Song',size=25,color='r')
plt.xlabel('Popularity',size=25,color='r')
plt.xticks(size=25)
plt.yticks(size=25)
plt.show()
```

## Less Popular Song

```python
less_popularity=df_tracks.set_index('name',inplace=False)
less_pop=less_popularity.sort_values(by='popularity',ascending=True)
less_pop.popularity.head(5)
```

## Most Loudest Song

```python
loudest=df_tracks.sort_values(by="loudness",ascending=False).head(6)
plt.figure(figsize=(15,10))
plt.bar(x=loudest.name,height=loudest.loudness)
for x,y in zip (loudest.name,loudest.loudness):
    plt.text(x,y+.02,y,size=12)
plt.title("Loudest song",size=30)
plt.ylabel('Loudness',size=20,color='r')
plt.xlabel('Name of Song',size=20,color='r')
plt.xticks(size=12)
plt.yticks(size=15)
plt.show()
```

# Spotify Features

```python
df_features.info()
```

## Duration of the Songs in Diffrent Genres

```python
plt.figure(figsize=(10,12))
plt.title("Duration of the Songs in Diffrent Genres",size=25)
sbn.barplot(y='genre',x='duration_ms',data=df_features)
plt.xlabel("Duration in Milli Seconds",size=20)
plt.ylabel("Genres",size=20)
```

## Top 5 Genre by Popularity

```python
famous=df_features.sort_values('popularity',ascending=False).head(10)
plt.figure(figsize=(10,8))
plt.title("",size=20)
sbn.barplot(y='genre',x='popularity',data=famous)
plt.ylabel("Genre",size=18)
plt.xlabel("Popularity",size=18)
```
